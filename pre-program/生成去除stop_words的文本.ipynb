```python
from Config import *
from HotelModel import *
import fileUtil
import re
import os
```

```python
basic_dir = 'dataset'
file_name = 'hotel_desc_seed.csv'
stop_file_name = 'stopwords.txt'

if os.path.exists(os.path.join(basic_dir,stop_file_name)):
    stops = fileUtil.read_data_as_array(basic_dir, stop_file_name)
else:
    print('file not exist')
stop_list = set(stops)
```

```python
print(stop_list)
```

```python
documents = []
with open(os.path.join(basic_dir,file_name), 'r', encoding='utf-8') as f:
    while True:
        line = f.readline()
        if len(line.strip()) != 0:
            #去除（）及内部的内容
            line = re.sub('\\（.*?\\）','',line.strip()).lower().replace('“', '').replace('”', '')
            for char in stop_list:
                line = line.replace(char, "")
            line = line.replace('。。', "END").replace('。', "END").replace('！', "END").replace('？', "END")
            documents.append(line)
        if not line:
            break
```

```python
print(documents[:10])
```

```python
def write_hotels_to_file(basic_dir, file_name, hotels_dict):
    with open(os.path.join(basic_dir,file_name), 'w+') as f:
        for item in  hotels_dict:
            f.write(item+'\n')
    f.close()
    print('Write hotel info to file done. ')
```

```python
write_hotels_to_file(basic_dir, 'hotel_desc_remove_stop_words.csv', documents)
```
